## Importing libraries

```python
%matplotlib inline
from sklearn.datasets import fetch_mldata 
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
```

## Importing MNIST data

```python
mnist = fetch_mldata('MNIST original')
```

```python
X, y = mnist['data'], mnist['target']
shuffle_index = np.random.permutation(70000)
X, y = X[shuffle_index] , y[shuffle_index]
```

## Selecting 2000 random instances to visulize

```python
X, y =  X[:2000], y[:2000]      # visulizing only 1/30 of the total data
```

# Using PCA for dimentionality reduction

```python
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)
```

```python
plt.figure(figsize = (8,6))
plt.scatter(X_reduced[:,0],X_reduced[:,1],c=y,cmap = 'jet')
plt.axis('off')
plt.colorbar()
```

# Using Kernel PCA

```python
from sklearn.decomposition import KernelPCA
rbf_pca = KernelPCA(n_components=2, kernel='rbf' , gamma=0.04)
X_reduced = rbf_pca.fit_transform(X)
```

```python
plt.figure(figsize = (8,6))
plt.scatter(X_reduced[:,0],X_reduced[:,1],c=y,cmap = 'jet')
plt.axis('off')
plt.colorbar()
```

# Using LLE Dimentionality reduction technique

```python
from sklearn.manifold import LocallyLinearEmbedding
lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10)
X_reduced = lle.fit_transform(X)
```

```python
plt.figure(figsize = (8,6))
plt.scatter(X_reduced[:,0],X_reduced[:,1],c=y,cmap = 'jet')
plt.axis('off')
plt.colorbar()
```

# Using t-SNE technique 

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=42)
X_reduced = tsne.fit_transform(X)
```

```python
plt.figure(figsize = (8,6))
plt.scatter(X_reduced[:,0],X_reduced[:,1],c=y,cmap = 'jet')
plt.axis('off')
plt.colorbar()
```

# Using Isomap

```python
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
X_reduced = iso.fit_transform(X)
```

```python
plt.figure(figsize = (8,6))
plt.scatter(X_reduced[:,0],X_reduced[:,1],c=y,cmap = 'jet')
plt.axis('off')
plt.colorbar()
```

# Using Multidimensional Scaling (MDS)

```python
from sklearn.manifold import MDS
mds = MDS(n_components=2, random_state=42)
X_reduced= mds.fit_transform(X)
```

```python
plt.figure(figsize = (8,6))
plt.scatter(X_reduced[:,0],X_reduced[:,1],c=y,cmap = 'jet')
plt.axis('off')
plt.colorbar()
```

# Using Linear Discriminant Analysis (LDA)

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components=2)
X_reduced = lda.fit_transform(X, y)
```

```python
plt.figure(figsize = (8,6))
plt.scatter(X_reduced[:,0],X_reduced[:,1],c=y,cmap = 'jet')
plt.axis('off')
plt.colorbar()
```

# Analysis:

It gives clusters having the same digit whereas in others different digits are overlapping on each other.
